MODELADO DE DATOS

```python
x = df_ohe.drop(['Sleep Disorder_Insomnia', 'Sleep Disorder_Sleep Apnea', 'Sleep Disorder_None'], axis=1)
y = df_ohe[['Sleep Disorder_Insomnia', 'Sleep Disorder_Sleep Apnea', 'Sleep Disorder_None']].idxmax(axis=1)
y = y.map({'Sleep Disorder_None': 0, 'Sleep Disorder_Insomnia': 1, 'Sleep Disorder_Sleep Apnea': 2})
x.head(5), y.head(5)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

MODELO ML: XGBoost

```python
model_xgb = XGBClassifier(
    use_label_encoder=False)

param_grid = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 7],
    'min_child_weight': [1, 3, 5],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'gamma': [0, 0.1, 0.2]
}

grid_search = GridSearchCV(
    estimator=model_xgb,
    param_grid=param_grid,
    cv=3, scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
y_pred = grid_search.best_estimator_.predict(x_test)
y_pred
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

```python
cm_df = pd.DataFrame(conf_matrix)

chart = alt.Chart(cm_df.reset_index().melt(id_vars=["index"], var_name="Predicted", value_name="Count")).mark_rect().encode(
    x='Predicted:N',
    y='index:N',
    color='Count:Q',
    tooltip=['index', 'Predicted', 'Count']
).properties(
    title="Confusion Matrix",
    width=400,
    height=400,
    
)

chart
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
classification_report(y_test, y_pred)
```
